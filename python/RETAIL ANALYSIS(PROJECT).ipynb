# Customers

```python
import pandas as pd


df = pd.read_csv("C:/Users/hp/Downloads/ExcellenC(PROJECT--1)/Ratails Analytics/customers.csv")

df.drop_duplicates (inplace=True)


df['gender'] = df['gender'].fillna('Unknown')


df['region'] = df['region'].fillna("unknown")


df['age'] = df['age'].fillna(df['age'].median())


df['first_name'] = df['first_name'].str.strip().str.title()

df['last_name'] = df['last_name'].str.strip().str.title()


df['gender'] = df['gender'].str.strip().str.title()

df['region'] = df['region'].str.strip().str.title()

df['signup_date'] = pd.to_datetime(df['signup_date'], errors='coerce')

df['age'] = df ['age'].astype(int)

df.to_csv(path_or_buf = "customers_cleaned.csv", index=False)

print("Data cleaned and saved successfully")
```

# Products

```python
import pandas as pd

df = pd.read_csv("C:/Users/hp/Downloads/ExcellenC(PROJECT--1)/Ratails Analytics/products.csv")

# 1. Drop duplicate rows
df.drop_duplicates(inplace=True)

# 2. Handle Missing Values: Fill missing 'brand' values with 'Unknown'
df['brand'] = df['brand'].fillna('Unknown')

# 3. Standardize Text Columns
# Standardize 'category'
df['category'] = df['category'].str.strip().str.title()

# Standardize 'brand' (strip whitespace)
df['brand'] = df['brand'].str.strip()

# 4. Feature Cleaning: Remove the brand prefix from product_name
# (e.g., "BrandA Shoulder" becomes "Shoulder")
df['product_name'] = df['product_name'].apply(lambda x: ' '.join(str(x).split(' ')[1:])).str.strip().str.title()

# Save the cleaned data to a new CSV file
df.to_csv(path_or_buf="products_cleaned.csv", index=False)

print("Products data cleaned and saved successfully ")
```

# Returns

```python
import pandas as pd

# Load the returns CSV file.
df = pd.read_csv("C:/Users/hp/Downloads/ExcellenC(PROJECT--1)/Ratails Analytics/returns.csv")

# 1. Drop duplicate rows
df.drop_duplicates(inplace=True)

# 2. Data Type Conversion
# Convert 'return_date' to datetime
df['return_date'] = pd.to_datetime(df['return_date'], errors='coerce')

# 3. Standardize Text Columns and IDs
df['return_id'] = df['return_id'].str.strip()
df['order_id'] = df['order_id'].str.strip()

# Standardize 'return_reason' (strip whitespace and title case)
df['return_reason'] = df['return_reason'].str.strip().str.title()

# Save the cleaned data to a new CSV file
# NOTE: Using 'returns_cleaned_v2.csv' to avoid the PermissionError from earlier
df.to_csv(path_or_buf="returns_cleaned1.csv", index=False)

# CORRECTED: Added the missing closing quote for the print statement
print("Returns data Cleaned and saved successfully")
```

# Sales-data

```python
import pandas as pd

# Load the sales_data CSV file.
df = pd.read_csv("C:/Users/hp/Downloads/ExcellenC(PROJECT--1)/Ratails Analytics/sales_data.csv")

# 1. Drop duplicate rows
df.drop_duplicates(inplace=True)

# 2. Handle Missing Values
# Fill missing 'store_id' values with 'Online', as these correspond to online sales
df['store_id'] = df['store_id'].fillna('Online')

# 3. Data Type Conversion and Standardization
# Convert 'order_date' to datetime
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')

# Standardize 'sales_channel'
df['sales_channel'] = df['sales_channel'].str.strip().str.title()

# 4. Ensure transactional IDs are clean (strip whitespace)
df['order_id'] = df['order_id'].str.strip()
df['customer_id'] = df['customer_id'].str.strip()
df['product_id'] = df['product_id'].str.strip()
df['store_id'] = df['store_id'].str.strip() # Re-apply strip after filling NaNs

# Save the cleaned data to a new CSV file
df.to_csv(path_or_buf="sales_cleaned.csv", index=False)

print("Sales data cleaned and saved successfully ")
```

# Stores

```python
import pandas as pd

# Load the stores CSV file.
df = pd.read_csv("C:/Users/hp/Downloads/ExcellenC(PROJECT--1)/Ratails Analytics/stores.csv")

# 1. Drop duplicate rows
df.drop_duplicates(inplace=True)

# 2. Standardize Text Columns
# Ensure 'store_id' is clean
df['store_id'] = df['store_id'].str.strip()

# Standardize text fields (strip whitespace and apply title case)
df['store_name'] = df['store_name'].str.strip().str.title()
df['store_type'] = df['store_type'].str.strip().str.title()
df['region'] = df['region'].str.strip().str.title()
df['city'] = df['city'].str.strip().str.title()

# Save the cleaned data to a new CSV file
df.to_csv(path_or_buf="stores_cleaned.csv", index=False)

print("Stores data cleaned and saved successfully as stores_cleaned.csv")
```
